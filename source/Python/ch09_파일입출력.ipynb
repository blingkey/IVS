# **1절 파일에 데이터를 저장하고 불러오기**

양방향 스트림은 불가. only 읽기 전용 or only 쓰기 전용

```python
# w : 있는 파일은 기존의 파일에 덮어쓰고 없는 파일은 생성 (쓰기 전용, 덮어쓰기)
# a : 있는 파일은 추가/ 없는 파일은 생성 (쓰기 전용, 추가하기)
# r : 읽기 전용

# 현재 ch09.txt 파일 없음 ∴ 생성해주기
f = open('data/ch09.txt', 'w') # encoding 기본값 CP949
print(f.writable())
```

```python
f.write('Hello\nWorld\nHello\nPython\n반갑습니다\n') # 한글은 CP949
f.write('여러번 write 가능\n') # 한글은 CP949
f.close()
```

```python
with open('data/ch09.txt', 'r') as f:
    data = f.readlines()
    print(data)
```

```python
with open('data/ch09.txt', 'w') as f:
    print('Python is easy\n', file=f)
```

**<font color = 'blue'>data/ch09_member1.csv 받기</font>**

```python
# UTF-8로 인코딩 된 CSV 파일을 read
with open('data/ch09_member1.csv', 'r', encoding='utf-8') as f:
    data = f.readlines()
    for row in data:
        print(row, end = "")
```
