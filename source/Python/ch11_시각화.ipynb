# 1절 시각화 개요
- 시각화 라이브러리 : matplotlib, seaborn, folium(지도 시각화)
- matplotlib : Pandas 데이터프레임 내부적으로 matplotlib 내장
- seaborn : matplotlib 기반으로 다양한 색, 테마, 차트 등을 추가한 라이브러리

```python
# 시작전 설정
import matplotlib.pyplot as plt
%matplotlib inline
# 그래프 해상도 높임
%config InlineBackend.figure_format='retina'
#한글설정
plt.rc('font', family='Malgun Gothic') #window os
#plt.rc('font', family='AppleGothic') #mac os
plt.rc('axes', unicode_minus=False)
# 경고 메세지 안보이게
import warnings
warnings.filterwarnings(action='ignore')
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.shape
```

```python
iris.plot(kind='line', figsize=(12,4))
plt.show()
```

```python
iris.plot(kind='line', figsize=(12,4))
# Put the legend out of the figure
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
iris.plot(kind='box', figsize=(12,4))
```
